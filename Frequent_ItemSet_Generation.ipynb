```python
#Import required libraries
import pandas as pd
import numpy as np
import itertools
```

```python
#Read the data from csv
mba_df=pd.read_csv("data1.csv")
```

```python
#Head of the mba dataframe
mba_df.head()
```

```python
#Considering the first two hundred records
mba_df=mba_df[0:200]
```

```python
#Handling of nan values and droping item(s) column
mba_df=mba_df.drop(columns={'Item(s)'})
```

```python
#Handling nan values
mba_df=mba_df.fillna(0)
```

```python
#Head of new dataframe after filling nan with 0
mba_df.head()
```

```python
#Generating the single item set
one_itemset=list()
row_count,column_count=mba_df.shape
for i in range(row_count):
    for j in range(column_count):
        if mba_df.iloc[i][j] is not 0:
             if mba_df.iloc[i][j] not in one_itemset: #Unique items not in list is appended
                    one_itemset.append(mba_df.iloc[i][j])
```

```python
#Count of number of unique itemset
len(one_itemset)
```

```python
#Calculating the count of single item set
single_itemset={}
#Intializing all the count to zero
for item in one_itemset:
    single_itemset[item]=0
for i in range(row_count):
    for j in range(column_count):
        if mba_df.iloc[i][j] is not 0:
              single_itemset[mba_df.iloc[i][j]]+=1
```

```python
#Defining minimum support count
min_sup=4
infrqt_items=list()
#eleminating all the items less than the support count i.e candidate pruning
for key,value in single_itemset.items():
    if value <= min_sup:#If candidate count is less than count then the item is removed
        infrqt_items.append(key)
for key in infrqt_items:
    del single_itemset[key]
```

```python
single_df=pd.DataFrame()
single_df['Items']=single_itemset.keys()
single_df['Count']=single_itemset.values()
```

```python
#Final single frequent item set
single_df
```

```python
#Generate all posiible two frequent item set
twoitem_df=pd.DataFrame(list(itertools.combinations(single_df['Items'], 2)))
```

```python
#Renaming the columns to respective values
twoitem_df=twoitem_df.rename(columns={0:'Item1',1:'Item2'})
```

```python
#Candidte two item set
twoitem_df
```

```python
#Initailizing the dict to generate the count of two candidate itemset
twoitem_set={}
for i in range(len(twoitem_df)):
    twoitem_set[i]=0
```

```python
#Checking the count of two item set
for key in twoitem_set.keys():
    for j in range(len(mba_df)):
             if set(twoitem_df.loc[key]).issubset(set(mba_df.loc[j])):
                twoitem_set[key]+=1
                    
```

```python
#Count of all transactions,i.e transaction or index mapped with count
twoitem_set
```

```python
#Pruning of itemset with count less than min supcount
infrqt_items=[]
min_sup=4
for key,value in twoitem_set.items():
    if value < min_sup:
        infrqt_items.append(key)
for key in infrqt_items:
    del twoitem_set[key]
```

```python
#Two itemset with index and count
twoitem_set
```

```python
#Creating of two frequent itemset
two_df=pd.DataFrame()
for index in twoitem_set.keys():
    row = twoitem_df.loc[ [index] ,: ]
    two_df=two_df.append(row)

    
```

```python
#Two frequent itemset
two_df
```

```python
#Creating three item set using Fk-1*F1 method
threeitem_set=[]
for key in twoitem_set.keys():
    for j in range(len(single_df)):
        subList=[]
        if(two_df.loc[key].Item1.find(single_df.loc[j].Items)==-1 and two_df.loc[key].Item2.find(single_df.loc[j].Items)==-1):
            subList=[two_df.loc[key].Item1,two_df.loc[key].Item2,single_df.loc[j].Items]
            threeitem_set.append(subList)
```

```python
#Three itemset
three_df=pd.DataFrame(threeitem_set)
```

```python
#Three itemset
three_df
```

```python
threeitem_set={}
for i in range(len(three_df)):
    threeitem_set[i]=0
```

```python
#Checking the count of three item set
for key in threeitem_set.keys():
    for j in range(len(mba_df)):
             if set(three_df.loc[key]).issubset(set(mba_df.loc[j])):
                threeitem_set[key]+=1
```

```python
#Three itemset count
threeitem_set
```

```python
#Pruning of three frequent item set with min support count
infrqt_items=[]
min_sup=4
for key,value in threeitem_set.items():
    if value < min_sup:
        infrqt_items.append(key)
for key in infrqt_items:
    del threeitem_set[key]
```

```python
#Final three frequent set
threeitem_set
```
